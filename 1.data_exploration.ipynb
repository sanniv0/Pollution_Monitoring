## Data Exploration (EDA)

Here we explore the dataset to understand its structure, missing values, and distributions.

```python

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load the dataset
file_path = "Dataset_Cities.csv"
df = pd.read_csv(file_path)

# Display first few rows
df.head()
```

```python

# Basic info about dataset
df.info()

# Check for missing values
df.isnull().sum().head(20)

# Summary statistics
df.describe().T.head(20)
```

```python

# Interactive plots with Plotly
import plotly.express as px

# Select key pollutants
pollutants = ['pollutant_id', 'pollutant_min', 'pollutant_max', 'pollutant_avg']

# Interactive histogram for each pollutant
for pollutant in pollutants:
    fig = px.histogram(df, x=pollutant, nbins=50, title=f"Distribution of {pollutant}", marginal="box")
    fig.show()
```

```python

# Correlation heatmap (numerical features)
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
df.corr(numeric_only=True).style.background_gradient(cmap="coolwarm")
```
